# Parallel Reader: Original Language + KJV + ESV, Side by Side

A close-reading companion to [`original_language_stylometry.ipynb`](original_language_stylometry.ipynb):
instead of running statistics across a whole book, this notebook renders a handful of key
passages verse-by-verse — original Hebrew/Greek word, transliteration, and gloss, alongside the
full KJV and (if you've set `ESV_API_KEY`) full ESV verse text.

**Setup:** sign up for a free ESV API key at https://api.esv.org and set it as the `ESV_API_KEY`
environment variable before running this notebook. Without it, the reader still works for
original + KJV; the ESV column just says so. ESV text is fetched one verse at a time and cached
in `data/esv_cache.json` — never bulk-downloaded (see `src/reader.py` docstring for why: Crossway's
free-API terms cap non-commercial use at 500 verses and forbid reproducing a whole book).

See [`docs/study_guide.md`](../docs/study_guide.md) for the reading path this notebook plugs into.

## Romans 1:1 vs the Pastorals / Hebrews — Paul's Greek opening formula

Paul's undisputed-letter opening (Romans) as a baseline before comparing it to Hebrews, whose
anonymous author never opens this way at all — one of the more basic reasons (independent of
stylometry) that Hebrews is not traditionally read as one of Paul's own letters.

```python
rd.print_parallel_table("Romans", 1, [1])
```

## Hebrews 1:1-2 — the polished opening period

Hebrews opens with an elaborate, rhetorically balanced Greek period quite unlike Paul's
compact epistolary formula above — a stylistic observation already made by Origen and Clement
of Alexandria, long before any stylometry (see `docs/background.md`).

```python
rd.print_parallel_table("Hebrews", 1, [1, 2])
```

## John 1:1 vs Revelation 1:1 — the Johannine style gulf

The Gospel's elegant, simple periodic Greek vs. Revelation's rougher, more paratactic style —
the same contrast the `original_language_stylometry.ipynb` notebook found (near-perfect
statistical separation, p=0.0002).

```python
rd.print_parallel_table("John", 1, [1])
rd.print_parallel_table("Revelation", 1, [1])
```

## 1 Peter 1:1 vs 2 Peter 1:1 — the sharpest stylistic gap, and the weakest evidence

These two letters show the largest raw stylistic distance of any pair in this whole project,
yet fail to reach statistical significance (see README) simply because 2 Peter is too short
(~1,100 Greek words) to give a permutation test any power. Read the Greek here and judge the
"baroque, almost vulgar... pretentiousness" Richard Bauckham describes (`docs/background.md`)
for yourself.

```python
rd.print_parallel_table("1Peter", 1, [1])
rd.print_parallel_table("2Peter", 1, [1])
```
